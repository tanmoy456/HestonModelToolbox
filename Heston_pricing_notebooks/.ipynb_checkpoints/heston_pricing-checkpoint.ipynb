```python
import numpy as np
from scipy.integrate import quad
```

The Heston pricing formula is given by:

$$
C(S_t, V_t, t, T) = S_t P_1 - K e^{-r(T-t)} P_2
$$

where the probabilities $P_j$ for $j=1,2$ are computed by inverse Fourier transform:

$$
P_j(x, V_t, T, K) = \frac{1}{2} + \frac{1}{\pi} \int_0^\infty \mathrm{Re} \left[ \frac{e^{-i \phi \ln K} f_j(x, V_t, T, \phi)}{i \phi} \right] d \phi
$$

with 
$$
x = \ln(S_t) 
$$

and

$$
f_j(x, V_t, T, \phi) = e^{C(\tau, \phi) + D(\tau, \phi) V_t + i \phi x}
$$

where $\tau = T - t$, and

$$
C(\tau, \phi) = r \phi i \tau + \frac{a}{\sigma^2} \left[ (b_j - \rho \sigma \phi i + d) \tau - 2 \ln \left( \frac{1 - g e^{d \tau}}{1 - g} \right) \right]
$$

$$
D(\tau, \phi) = \frac{b_j - \rho \sigma \phi i + d}{\sigma^2} \cdot \frac{1 - e^{d \tau}}{1 - g e^{d \tau}}
$$

$$
g = \frac{b_j - \rho \sigma \phi i + d}{b_j - \rho \sigma \phi i - d}
$$

$$
d = \sqrt{(\rho \sigma \phi i - b_j)^2 - \sigma^2 (2 u_j \phi i - \phi^2)}
$$

for $j = 1, 2$, with parameters

$$
u_1 = \frac{1}{2}, \quad u_2 = -\frac{1}{2}
$$

$$
a = \kappa \theta, \quad b_1 = \kappa + \lambda - \rho \sigma, \quad b_2 = \kappa + \lambda
$$


```python
# Model parameters
S0 = 100.0    # Initial stock price
K = 100.0     # Strike price
r = 0.05      # Risk-free rate
T = 1.0       # Time to maturity
kappa = 2.0   # Mean reversion rate
theta = 0.05  # Long-term average volatility
sigma = 0.3   # Volatility of volatility
rho = -0.5    # Correlation coefficient
v0 = 0.05     # Initial volatility
```

```python
# Define characteristic functions
def heston_characteristic_function(u, S0, K, r, T, kappa, theta, sigma, rho, v0):
   xi = kappa - rho * sigma * 1j * u
   d = np.sqrt((rho * sigma * 1j * u - xi)**2 - sigma**2 * (-u * 1j - u**2))
   g = (xi - rho * sigma * 1j * u - d) / (xi - rho * sigma * 1j * u + d)
   C = r * 1j * u * T + (kappa * theta) / sigma**2 * ((xi - rho * sigma * 1j * u - d) * T - 2 * np.log((1 - g * np.exp(-d * T)) / (1 - g)))
   D = (xi - rho * sigma * 1j * u - d) / sigma**2 * ((1 - np.exp(-d * T)) / (1 - g * np.exp(-d * T)))
   return np.exp(C + D * v0 + 1j * u * np.log(S0))

# Define functions to compute call and put options prices
def heston_call_price(S0, K, r, T, kappa, theta, sigma, rho, v0):
   integrand = lambda u: np.real(np.exp(-1j * u * np.log(K)) / (1j * u) * heston_characteristic_function(u - 1j, S0, K, r, T, kappa, theta, sigma, rho, v0))
   integral, _ = quad(integrand, 0, np.inf)
   return np.exp(-r * T) * 0.5 * S0 - np.exp(-r * T) / np.pi * integral


def heston_put_price(S0, K, r, T, kappa, theta, sigma, rho, v0):
   integrand = lambda u: np.real(np.exp(-1j * u * np.log(K)) / (1j * u) * heston_characteristic_function(u - 1j, S0, K, r, T, kappa, theta, sigma, rho, v0))
   integral, _ = quad(integrand, 0, np.inf)
   return np.exp(-r * T) / np.pi * integral - S0 + K * np.exp(-r * T)
```

```python
# Calculate call and put option prices
call_price = heston_call_price(S0, K, r, T, kappa, theta, sigma, rho, v0)
put_price = heston_put_price(S0, K, r, T, kappa, theta, sigma, rho, v0)


print("European Call Option Price:", np.round(call_price, 2))
print("European Put Option Price:", np.round(put_price, 2))
```

### Heston Model Characteristic Function — Analytical Formula

The characteristic function $ f_j $ is defined as:

$$
f_j(x, V_t, T, \phi) = \exp \left[ C(T - t, \phi) + D(T - t, \phi) \, V_t + i \phi x \right]
$$

where:

$$
x = \ln(S_t)
$$

and $ \phi $ is the integration variable (denoted as $ u $ in code).

---

#### **Intermediate variables:**

$$
d = \sqrt{(\rho \sigma i \phi - b_j)^2 - \sigma^2 \left( 2 u_j i \phi - \phi^2 \right)}
$$

$$
g = \frac{b_j - \rho \sigma i \phi + d}{b_j - \rho \sigma i \phi - d}
$$

---

#### **C and D terms:**

$$
C(\tau, \phi) = r \, \phi i \, \tau
+ \frac{a}{\sigma^2} \left[ (b_j - \rho \sigma i \phi + d) \, \tau
- 2 \ln \left( \frac{1 - g \, e^{d \tau}}{1 - g} \right) \right]
$$

$$
D(\tau, \phi) = \frac{b_j - \rho \sigma i \phi + d}{\sigma^2}
\cdot \frac{1 - e^{d \tau}}{1 - g \, e^{d \tau}}
$$

where $ \tau = T - t $.

---

#### **Parameter definitions:**

$$
u_1 = \frac{1}{2}, \quad u_2 = -\frac{1}{2}
$$

$$
a = \kappa \theta
$$

$$
b_1 = \kappa + \lambda - \rho \sigma, \quad b_2 = \kappa + \lambda
$$


```python
def heston_characteristic_function(u, S0, K, r, T, kappa, theta, sigma, rho, v0):
    # This corresponds to the term b_j - rho*sigma*phi*i in the formulas.
    # The variable 'u' in the code represents the integration variable 'phi'.
    xi = kappa - rho * sigma * 1j * u

    # This calculates d = sqrt((rho*sigma*phi*i - b_j)^2 - sigma^2*(2*u_j*phi*i - phi^2))
    # The code uses a simplified but equivalent formulation.
    d = np.sqrt((rho * sigma * 1j * u - xi)**2 - sigma**2 * (-u * 1j - u**2))

    # This calculates g = (b_j - rho*sigma*phi*i + d) / (b_j - rho*sigma*phi*i - d)
    g = (xi - rho * sigma * 1j * u - d) / (xi - rho * sigma * 1j * u + d)

    # This calculates the C(T-t, phi) term.
    C = r * 1j * u * T + (kappa * theta) / sigma**2 * ((xi - rho * sigma * 1j * u - d) * T - 2 * np.log((1 - g * np.exp(-d * T)) / (1 - g)))

    # This calculates the D(T-t, phi) term.
    D = (xi - rho * sigma * 1j * u - d) / sigma**2 * ((1 - np.exp(-d * T)) / (1 - g * np.exp(-d * T)))

    # This combines all the pieces to return the final value of the characteristic function f_j.
    return np.exp(C + D * v0 + 1j * u * np.log(S0))
```

### Heston Model European Call Option Price — Analytical Formula

The price of a European call option under the Heston stochastic volatility model is:

$$
C(S_t, V_t, t, T) = S_t \, P_1 - K e^{-r(T-t)} P_2
$$

where

- $S_t$ is the spot (underlying asset) price at time $t$,
- $K$ is the strike price,
- $r$ is the risk-free rate,
- $V_t$ is the variance at time $t$,
- $T$ is the maturity time,
- $P_1$ and $P_2$ are risk-neutral probabilities for exercising the option.

---

#### Risk-Neutral Probabilities from the Characteristic Function

The probabilities $P_j$ (for $j = 1, 2$) are computed by integrating the characteristic function $f_j$:

$$
P_j(x, V_t, T, K) = \frac{1}{2} + \frac{1}{\pi} \int_0^{\infty} \mathrm{Re} \left[ \frac{e^{-i \phi \ln(K)} f_j(x, V_t, T, \phi)}{i \phi} \right] d\phi
$$

where

- $x = \ln(S_t)$
- $\phi$ is the integration variable (Fourier transform variable),
- $f_j(x, V_t, T, \phi)$ is the Heston model characteristic function.

---

#### Code Explanation

The code implements this pricing by numerically integrating the characteristic function directly using the above formula, combining the calculations for $P_1$ and $P_2$ to get the call price:

1. Evaluates the characteristic function $f_j$ for each $\phi$.
2. Integrates over $\phi$ from $0$ to $\infty$.
3. Computes $P_1$ and $P_2$ using the result.
4. Calculates the final call price using $C = S_t \, P_1 - K e^{-r(T-t)} P_2$.

This direct-integration approach is common in quantitative finance libraries for numerically computing option prices under the Heston model.

---


```python
def heston_call_price(S0, K, r, T, kappa, theta, sigma, rho, v0):
    # Defines the function to be integrated, which is derived from the formula for P_2.
    # The term heston_characteristic_function(u - 1j, ...) is related to f_2.
    integrand = lambda u: np.real(np.exp(-1j * u * np.log(K)) / (1j * u) * heston_characteristic_function(u - 1j, S0, K, r, T, kappa, theta, sigma, rho, v0))

    # Numerically computes the integral from 0 to infinity.
    integral, _ = quad(integrand, 0, np.inf)

    # This is a rearranged version of the standard call price formula.
    return np.exp(-r * T) * 0.5 * S0 - np.exp(-r * T) / np.pi * integral
```

### Heston Model European Put Option Price — Analytical Formula

The price of a European put option under the Heston stochastic volatility model can be obtained using the **put–call parity relationship**:

$$
P = C - S_0 + K e^{-rT}
$$

where

- $P$ is the put price,
- $C$ is the call price (computed via the Heston model),
- $S_0$ is the current spot price,
- $K$ is the strike price,
- $r$ is the risk-free interest rate,
- $T$ is the time to maturity.

---

#### Alternative Direct Integration

Alternatively, a direct integration formula similar to that for the call price can be derived from the characteristic function:

$$
P(S_0, K) = \frac{e^{-rT}}{\pi} \int_{0}^{\infty} 
\mathrm{Re} \left[ \frac{e^{-i u \ln K} \, \varphi(u - i)}{i u} \right] du
- S_0 + K e^{-rT}
$$

where $\varphi(u)$ is the Heston model characteristic function.

---

#### Code Insight

The provided code does **not** separately integrate for the put price.  
Instead, it:

1. Computes the call option price using the Fourier inversion formula.
2. Applies the put–call parity formula $P = C - S_0 + K e^{-rT}$ to get the put price.

This approach is computationally efficient since it reuses the call price integration result.


```python
def heston_put_price(S0, K, r, T, kappa, theta, sigma, rho, v0):
    # The integrand and integral calculation are identical to the call price function.
    # This is because the integral is a common component for both.
    integrand = lambda u: np.real(np.exp(-1j * u * np.log(K)) / (1j * u) * heston_characteristic_function(u - 1j, S0, K, r, T, kappa, theta, sigma, rho, v0))
    integral, _ = quad(integrand, 0, np.inf)

    # This formula calculates the put price. It can be derived by substituting
    # the call price expression into the put-call parity formula.
    return np.exp(-r * T) / np.pi * integral - S0 + K * np.exp(-r * T)
```

```python
# Calculate call and put option prices
call_price = heston_call_price(S0, K, r, T, kappa, theta, sigma, rho, v0)
put_price = heston_put_price(S0, K, r, T, kappa, theta, sigma, rho, v0)


print("European Call Option Price:", np.round(call_price, 2))
print("European Put Option Price:", np.round(put_price, 2))
```

```python
import numpy as np
from scipy import integrate

def heston_cf(phi, j, S0, v0, r, tau, kappa, theta, sigma, rho, lam=0.0):
    if j == 1:
        uj = 0.5
        bj = kappa + lam - rho * sigma
    elif j == 2:
        uj = -0.5
        bj = kappa + lam
    else:
        raise ValueError("j must be 1 or 2")
    
    i = 1j
    a = kappa * theta
    xi = bj - rho * sigma * i * phi
    d = np.sqrt((rho * sigma * i * phi - bj)**2 - sigma**2 * (2 * uj * i * phi - phi**2))
    g = (bj - rho * sigma * i * phi + d) / (bj - rho * sigma * i * phi - d)
    C = r * i * phi * tau + (a / sigma**2) * ((bj - rho * sigma * i * phi + d) * tau - 2.0 * np.log((1 - g * np.exp(d * tau)) / (1 - g)))
    D = (bj - rho * sigma * i * phi + d) / sigma**2 * ((1 - np.exp(d * tau)) / (1 - g * np.exp(d * tau)))
    x = np.log(S0)
    return np.exp(C + D * v0 + i * phi * x)

def P_j_integral(j, S0, v0, r, tau, K, kappa, theta, sigma, rho, lam=0.0):
    lnK = np.log(K)
    def integrand(phi):
        f = heston_cf(phi, j, S0, v0, r, tau, kappa, theta, sigma, rho, lam)
        val = np.real(np.exp(-1j * phi * lnK) * f / (1j * phi))
        return val
    result, _ = integrate.quad(integrand, 1e-8, 200.0, epsabs=1e-8, epsrel=1e-6, limit=200)
    return 0.5 + result / np.pi

def heston_call_price(S0, K, r, tau, kappa, theta, sigma, rho, v0, lam=0.0):
    P1 = P_j_integral(1, S0, v0, r, tau, K, kappa, theta, sigma, rho, lam)
    P2 = P_j_integral(2, S0, v0, r, tau, K, kappa, theta, sigma, rho, lam)
    return S0 * P1 - K * np.exp(-r * tau) * P2

def heston_put_price(S0, K, r, tau, kappa, theta, sigma, rho, v0, lam=0.0):
    # Put-call parity: P = C - S0 + K e^{-r tau}
    call = heston_call_price(S0, K, r, tau, kappa, theta, sigma, rho, v0, lam)
    return call - S0 + K * np.exp(-r * tau)


# Given parameters
S0 = 100.0    # Initial stock price
K = 100.0     # Strike price
r = 0.05      # Risk-free rate
T = 1.0       # Time to maturity
kappa = 2.0   # Mean reversion rate
theta = 0.05  # Long-term average volatility
sigma = 0.3   # Volatility of volatility
rho = -0.5    # Correlation coefficient
v0 = 0.05     # Initial volatility

call_price = heston_call_price(S0, K, r, T, kappa, theta, sigma, rho, v0)
put_price = heston_put_price(S0, K, r, T, kappa, theta, sigma, rho, v0)

print("European Call Option Price:", np.round(call_price, 2))
print("European Put Option Price:", np.round(put_price, 2))
```

```python
# Reproduce a numerically robust Heston implementation and show equivalence
import numpy as np
from scipy import integrate

# Parameters (from your example)
S0 = 100.0
K = 100.0
r = 0.05
T = 1.0
kappa = 2.0
theta = 0.05
sigma = 0.3
rho = -0.5
v0 = 0.05
lam = 0.0  # market price of volatility risk

def heston_cf_j(phi, j, S0, v0, r, tau, kappa, theta, sigma, rho, lam=0.0):
    """
    Heston characteristic function f_j(phi) (paper-style) for real or complex phi.
    j = 1 or 2 selects u_j and b_j.
    Returns complex value.
    Uses paper conventions with exp(d*tau) (so be consistent).
    """
    i = 1j
    if j == 1:
        uj = 0.5
        bj = kappa + lam - rho * sigma
    else:
        uj = -0.5
        bj = kappa + lam
    a = kappa * theta

    # ensure phi can be complex (np operations handle complex arrays/scalars)
    xi = bj - rho * sigma * i * phi
    d = np.sqrt((rho * sigma * i * phi - bj)**2 - sigma**2 * (2 * uj * i * phi - phi**2))
    g = (bj - rho * sigma * i * phi + d) / (bj - rho * sigma * i * phi - d)

    # Use branch-consistent log by relying on numpy; for more advanced control
    # one may enforce continuity across integration points.
    C = r * i * phi * tau + (a / sigma**2) * ((bj - rho * sigma * i * phi + d) * tau - 2.0 * np.log((1 - g * np.exp(d * tau)) / (1 - g)))
    D = (bj - rho * sigma * i * phi + d) / sigma**2 * ((1 - np.exp(d * tau)) / (1 - g * np.exp(d * tau)))
    x = np.log(S0)
    return np.exp(C + D * v0 + i * phi * x)

def Pj_by_integration(j, S0, v0, r, tau, K, kappa, theta, sigma, rho, lam=0.0):
    lnK = np.log(K)
    def integrand(phi):
        # integrand for P_j: Re[ e^{-i phi lnK} f_j(phi) / (i phi) ]
        f = heston_cf_j(phi, j, S0, v0, r, tau, kappa, theta, sigma, rho, lam)
        val = np.real(np.exp(-1j * phi * lnK) * f / (1j * phi))
        return val
    # integrate from small positive to finite upper
    res, err = integrate.quad(integrand, 1e-8, 200.0, epsabs=1e-8, epsrel=1e-6, limit=200)
    return 0.5 + res / np.pi

def call_by_P1P2(S0, K, r, tau, kappa, theta, sigma, rho, v0, lam=0.0):
    P1 = Pj_by_integration(1, S0, v0, r, tau, K, kappa, theta, sigma, rho, lam)
    P2 = Pj_by_integration(2, S0, v0, r, tau, K, kappa, theta, sigma, rho, lam)
    return S0 * P1 - K * np.exp(-r * tau) * P2, P1, P2

# Single-integral Lewis-style using phi - i
def phi_full(phi_complex):
    """Helper: characteristic function phi_X evaluated at complex argument"""
    # we'll re-use j=2 form for phi_X (characteristic of log-price under RN)
    # In previous derivations phi_X corresponds to f_2 (no S_T factor)
    return heston_cf_j(phi_complex, 2, S0, v0, r, T, kappa, theta, sigma, rho, lam)

def call_by_lewis_single(S0, K, r, tau, kappa, theta, sigma, rho, v0, lam=0.0):
    lnK = np.log(K)
    def integrand(u):
        # u is real integration variable; evaluate phi at u - i
        z = u - 1j
        val = np.exp(-1j * u * lnK) * phi_full(z) / (1j * u)
        return np.real(val)
    res, err = integrate.quad(integrand, 1e-8, 200.0, epsabs=1e-8, epsrel=1e-6, limit=200)
    return np.exp(-r * tau) * (0.5 * S0 - res / np.pi)

# Compute prices
call_p1p2, P1, P2 = call_by_P1P2(S0, K, r, T, kappa, theta, sigma, rho, v0, lam)
call_lewis = call_by_lewis_single(S0, K, r, T, kappa, theta, sigma, rho, v0, lam)
put_p1p2 = call_p1p2 - S0 + K * np.exp(-r * T)  # put-call parity

call_p1p2, call_lewis, put_p1p2, P1, P2

```

```python
# Full comparison: P1/P2, Lewis combined single-integral, COS, Carr-Madan FFT, and Monte Carlo (QE)
# Robustified handling to reduce divide-by-zero and branch issues where possible.
import numpy as np
from scipy import integrate, fftpack, stats
import time

# Model parameters
S0 = 100.0; K = 100.0; r = 0.05; T = 1.0
kappa = 2.0; theta = 0.05; sigma = 0.3; rho = -0.5; v0 = 0.05; lam = 0.0

# small eps to avoid divide-by-zero
_eps = 1e-16
i = 1j

def heston_cf_j(phi, j, S0, v0, r, tau, kappa, theta, sigma, rho, lam=0.0):
    """
    Heston characteristic function f_j(phi) (paper-style).
    Accepts scalar or numpy array phi (real or complex).
    """
    phi = np.asarray(phi)  # allow array
    # ensure complex dtype to avoid accidental float operations
    phi = phi.astype(np.complex128)
    if j == 1:
        uj = 0.5
        bj = kappa + lam - rho * sigma
    else:
        uj = -0.5
        bj = kappa + lam
    a = kappa * theta

    xi = bj - rho * sigma * 1j * phi
    # compute d using complex sqrt
    d = np.sqrt((rho * sigma * 1j * phi - bj)**2 - sigma**2 * (2 * uj * 1j * phi - phi**2))
    # avoid zero denominator in g by adding tiny eps
    denom = (bj - rho * sigma * 1j * phi - d)
    denom = np.where(np.abs(denom) == 0, denom + _eps, denom)
    g = (bj - rho * sigma * 1j * phi + d) / denom

    # similarly avoid problematic values inside log by regularizing argument
    exp_dt = np.exp(d * tau)
    with np.errstate(all='ignore'):
        log_arg = (1 - g * exp_dt) / (1 - g)
        # avoid negative/zero inside log for pathological complex values
        log_arg = np.where(np.abs(log_arg) == 0, log_arg + _eps, log_arg)
        C = r * 1j * phi * tau + (a / sigma**2) * ((bj - rho * sigma * 1j * phi + d) * tau - 2.0 * np.log(log_arg))
        D = (bj - rho * sigma * 1j * phi + d) / sigma**2 * ((1 - exp_dt) / (1 - g * exp_dt))
    x = np.log(S0)
    return np.exp(C + D * v0 + 1j * phi * x)


# 1) P1/P2 integration (Heston original two integrals)
def Pj_by_integration(j, S0, v0, r, tau, K, kappa, theta, sigma, rho, lam=0.0, upper=400.0):
    lnK = np.log(K)
    def integrand(phi):
        # phi is real positive; pass complex phi into CF
        f = heston_cf_j(phi, j, S0, v0, r, tau, kappa, theta, sigma, rho, lam)
        val = np.real(np.exp(-1j * phi * lnK) * f / (1j * phi + _eps))
        return val
    res, err = integrate.quad(integrand, 1e-8, upper, epsabs=1e-8, epsrel=1e-7, limit=200)
    return 0.5 + res / np.pi

def call_by_P1P2(S0, K, r, tau, kappa, theta, sigma, rho, v0, lam=0.0):
    P1 = Pj_by_integration(1, S0, v0, r, tau, K, kappa, theta, sigma, rho, lam)
    P2 = Pj_by_integration(2, S0, v0, r, tau, K, kappa, theta, sigma, rho, lam)
    return S0 * P1 - K * np.exp(-r * tau) * P2, P1, P2

# 2) Lewis combined single-integral (algebraic combination of P1 and P2)
def call_by_lewis_combined(S0, K, r, tau, kappa, theta, sigma, rho, v0, lam=0.0, upper=400.0):
    lnK = np.log(K)
    def integrand(u):
        z = u - 1j
        # phi_log is f2 (characteristic of log-price)
        phi_u_minus_i = heston_cf_j(z, 2, S0, v0, r, tau, kappa, theta, sigma, rho, lam)
        phi_u = heston_cf_j(u, 2, S0, v0, r, tau, kappa, theta, sigma, rho, lam)
        num = S0 * phi_u_minus_i - K * phi_u
        val = np.real(np.exp(-1j * u * lnK) * num / (1j * u + _eps))
        return val
    res, err = integrate.quad(integrand, 1e-8, upper, epsabs=1e-8, epsrel=1e-7, limit=200)
    C = np.exp(-r * tau) * (0.5 * S0 - 0.5 * K + res / np.pi)
    return C

# 3) COS method (Fang & Oosterlee) — implement with robust truncation using cumulants approx
def cos_price_call(S0, K, r, T, N=512, L=10):
    # characteristic function of log-price
    cf = lambda u: heston_cf_j(u, 2, S0, v0, r, T, kappa, theta, sigma, rho, lam)
    # approximate cumulants for truncation (use moment-matching approx)
    # crude approximations: mean and variance estimates for log-price
    c1 = np.log(S0) + (r - 0.5 * theta) * T  # approximate first cumulant
    c2 = (v0 + theta) * T  # crude variance approx
    a = c1 - L * np.sqrt(c2)
    b = c1 + L * np.sqrt(c2)
    x0 = np.log(S0)
    k = np.arange(N)
    u = k * np.pi / (b - a)
    # payoff coefficients for call on [a,b]
    def chi_psi(k):
        kpi = k * np.pi / (b - a)
        c = np.log(K)
        d = b
        with np.errstate(divide='ignore', invalid='ignore'):
            chi = (np.cos(kpi*(d-a))*np.exp(d) - np.cos(kpi*(c-a))*np.exp(c) +
                   kpi*(np.sin(kpi*(d-a))*np.exp(d) - np.sin(kpi*(c-a))*np.exp(c))) / (1 + kpi**2)
            psi = (np.sin(kpi*(d-a)) - np.sin(kpi*(c-a))) / kpi
        psi[0] = d - c
        return chi, psi
    chi, psi = chi_psi(k)
    Vk = 2.0/(b-a) * (chi - psi)
    # evaluate cf at u
    cf_vals = cf(u)
    temp = Vk * np.exp(1j * np.outer(u, x0)) * cf_vals
    price = np.exp(-r*T) * np.real(np.sum(temp))
    return price

# 4) Carr-Madan FFT implementation (with damping alpha)
def carr_madan_fft_price(S0, r, T, kappa, theta, sigma, rho, v0, K, alpha=1.5, N=4096, eta=0.125):
    # grid parameters
    lambd = 2*np.pi / (N * eta)
    b = np.log(S0) - N * lambd / 2
    v = eta * np.arange(N)
    lnS0 = np.log(S0)
    # modified characteristic function integrand psi(v)
    def psi(vj):
        phi = vj - (alpha + 1) * 1j
        val = heston_cf_j(phi, 2, S0, v0, r, T, kappa, theta, sigma, rho, lam)
        numer = np.exp(-r*T) * val
        denom = (alpha + 1j * vj) * (alpha + 1j * vj + 1)
        return numer / (denom + _eps)
    # Simpson weights
    w = np.ones(N)
    w[1::2] = 4
    w[2::2] = 2
    w *= eta/3.0
    fft_input = np.exp(1j * b * v) * psi(v) * w
    fft_vals = fftpack.fft(fft_input)
    ks = b + lambd * np.arange(N)
    strikes = np.exp(ks)
    Cvals = np.exp(-alpha * ks) / np.pi * np.real(fft_vals)
    # interpolate for desired K
    price = np.interp(K, strikes, Cvals)
    return price

# 5) Monte Carlo with Andersen QE + antithetic
def qe_mc_price(S0, K, r, T, kappa, theta, sigma, rho, v0, n_steps=200, n_paths=200000, seed=42):
    np.random.seed(seed)
    dt = T / n_steps
    half = n_paths // 2
    S = np.full(n_paths, S0)
    v = np.full(n_paths, v0)
    for t in range(n_steps):
        z1 = np.random.randn(half)
        z2 = np.random.randn(half)
        w1 = np.concatenate([z1, -z1])
        w2 = np.concatenate([z2, -z2])
        eps = np.sqrt(1 - rho**2) * w2
        dw1 = np.sqrt(dt) * w1
        dw2 = np.sqrt(dt) * (rho * w1 + eps)
        # QE for v
        m = theta + (v - theta) * np.exp(-kappa * dt)
        s2 = v * sigma**2 * np.exp(-kappa * dt) * (1 - np.exp(-kappa * dt)) / (kappa + _eps)
        phi = s2 / (m**2 + _eps)
        vnew = np.empty_like(v)
        mask = phi <= 1.5
        # first branch: non-central chi-sq approx via gamma
        if mask.any():
            beta = (1.5 / (phi[mask] + _eps) - 1.0) / 2.0
            alpha_q = m[mask] / (beta + _eps)
            vnew[mask] = alpha_q * stats.gamma.rvs(beta, scale=1.0, size=mask.sum())
        # second branch: exponential
        mask2 = ~mask
        if mask2.any():
            p = (phi[mask2] - 1.0) / (phi[mask2] + 1.0)
            beta2 = (1 - p) / (m[mask2] + _eps)
            u = np.random.rand(mask2.sum())
            vnew[mask2] = np.where(u <= p, 0.0, -np.log((1 - p) / (1 - u + _eps)) * beta2)
        vnew = np.maximum(vnew, 0.0)
        # asset update (log-Euler)
        S = S * np.exp((r - 0.5 * v) * dt + np.sqrt(v) * dw1)
        v = vnew
    payoff = np.maximum(S - K, 0.0)
    price = np.mean(np.exp(-r*T) * payoff)
    stderr = np.std(np.exp(-r*T) * payoff) / np.sqrt(n_paths)
    return price, stderr

# Run all methods and report results
results = {}
t0 = time.time()
call_p1p2, P1, P2 = call_by_P1P2(S0, K, r, T, kappa, theta, sigma, rho, v0, lam)
results['P1P2_call'] = (call_p1p2, time.time() - t0, P1, P2)

t0 = time.time()
call_lewis = call_by_lewis_combined(S0, K, r, T, kappa, theta, sigma, rho, v0, lam)
results['Lewis_combined'] = (call_lewis, time.time() - t0)

t0 = time.time()
call_cos = cos_price_call(S0, K, r, T, N=512, L=12)
results['COS_call'] = (call_cos, time.time() - t0)

t0 = time.time()
call_cm = carr_madan_fft_price(S0, r, T, kappa, theta, sigma, rho, v0, K, alpha=1.5, N=4096, eta=0.125)
results['CarrMadan_call'] = (call_cm, time.time() - t0)

t0 = time.time()
call_mc, mc_se = qe_mc_price(S0, K, r, T, kappa, theta, sigma, rho, v0, n_steps=200, n_paths=120000, seed=2025)
results['MC_QE_call'] = (call_mc, time.time() - t0, mc_se)

results

```

```python
import QuantLib as ql

# 1. Set evaluation date
today = ql.Date.todaysDate()
ql.Settings.instance().evaluationDate = today

# 2. Term structures
day_count = ql.Actual365Fixed()
risk_free_ts = ql.YieldTermStructureHandle(
    ql.FlatForward(today, 0.05, day_count)
)
dividend_ts = ql.YieldTermStructureHandle(
    ql.FlatForward(today, 0.0, day_count)
)

# 3. Spot price handle
spot_handle = ql.QuoteHandle(ql.SimpleQuote(100.0))

# 4. Define Heston process
heston_process = ql.HestonProcess(
    risk_free_ts,
    dividend_ts,
    spot_handle,
    0.05,    # v0
    2.0,     # kappa
    0.05,    # theta
    0.3,     # sigma
    -0.5     # rho
)

# 5. Model + Engine
heston_model = ql.HestonModel(heston_process)
engine = ql.AnalyticHestonEngine(heston_model)

# 6. European Call Option setup
strike = 100.0
maturity_date = today + int(365.0 * 1.0)  # T=1 year
payoff = ql.PlainVanillaPayoff(ql.Option.Call, strike)
exercise = ql.EuropeanExercise(maturity_date)
option = ql.VanillaOption(payoff, exercise)
option.setPricingEngine(engine)

# 7. Compute price
price = option.NPV()
print("QuantLib Heston Analytic Price: ", price)
```

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as st
import enum 

# This class defines puts and calls
class OptionType(enum.Enum):
    CALL = 1.0
    PUT = -1.0

def CallPutOptionPriceCOSMthd(cf,CP,S0,r,tau,K,N,L):
    # cf   - characteristic function as a functon, in the book denoted as \varphi
    # CP   - C for call and P for put
    # S0   - Initial stock price
    # r    - interest rate (constant)
    # tau  - time to maturity
    # K    - list of strikes
    # N    - Number of expansion terms
    # L    - size of truncation domain (typ.:L=8 or L=10)  
        
    # reshape K to a column vector
    if K is not np.array:
        K = np.array(K).reshape([len(K),1])
    
    #assigning i=sqrt(-1)
    i = complex(0.0,1.0) 
    x0 = np.log(S0 / K)   
    
    # truncation domain
    a = 0.0 - L * np.sqrt(tau)
    b = 0.0 + L * np.sqrt(tau)
    
    # sumation from k = 0 to k=N-1
    k = np.linspace(0,N-1,N).reshape([N,1])  
    u = k * np.pi / (b - a);  

    # Determine coefficients for Put Prices  
    H_k = CallPutCoefficients(CP,a,b,k)   
    mat = np.exp(i * np.outer((x0 - a) , u))
    temp = cf(u) * H_k 
    temp[0] = 0.5 * temp[0]    
    value = np.exp(-r * tau) * K * np.real(mat.dot(temp))     
    return value

def Chi_Psi(a,b,c,d,k):
    psi = np.sin(k * np.pi * (d - a) / (b - a)) - np.sin(k * np.pi * (c - a)/(b - a))
    psi[1:] = psi[1:] * (b - a) / (k[1:] * np.pi)
    psi[0] = d - c
    
    chi = 1.0 / (1.0 + np.power((k * np.pi / (b - a)) , 2.0)) 
    expr1 = np.cos(k * np.pi * (d - a)/(b - a)) * np.exp(d)  - np.cos(k * np.pi 
                  * (c - a) / (b - a)) * np.exp(c)
    expr2 = k * np.pi / (b - a) * np.sin(k * np.pi * 
                        (d - a) / (b - a))   - k * np.pi / (b - a) * np.sin(k 
                        * np.pi * (c - a) / (b - a)) * np.exp(c)
    chi = chi * (expr1 + expr2)
    
    value = {"chi":chi,"psi":psi }
    return value

# Determine coefficients for Put Prices 
def CallPutCoefficients(CP,a,b,k):
    if CP==OptionType.CALL:                  
        c = 0.0
        d = b
        coef = Chi_Psi(a,b,c,d,k)
        Chi_k = coef["chi"]
        Psi_k = coef["psi"]
        if a < b and b < 0.0:
            H_k = np.zeros([len(k),1])
        else:
            H_k      = 2.0 / (b - a) * (Chi_k - Psi_k)  
    elif CP==OptionType.PUT:
        c = a
        d = 0.0
        coef = Chi_Psi(a,b,c,d,k)
        Chi_k = coef["chi"]
        Psi_k = coef["psi"]
        H_k      = 2.0 / (b - a) * (- Chi_k + Psi_k)                  
    return H_k    

def ChFHestonModel(r,tau,kappa,gamma,vbar,v0,rho):
    i = complex(0.0,1.0)
    D1 = lambda u: np.sqrt(np.power(kappa-gamma*rho*i*u,2)+(u*u+i*u)*gamma*gamma)
    g  = lambda u: (kappa-gamma*rho*i*u-D1(u))/(kappa-gamma*rho*i*u+D1(u))
    C  = lambda u: (1.0-np.exp(-D1(u)*tau))/(gamma*gamma*(1.0-g(u)*np.exp(-D1(u)*tau)))\
        *(kappa-gamma*rho*i*u-D1(u))
    # Note that we exclude the term -r*tau, as the discounting is performed in the COS method
    A  = lambda u: r * i*u *tau + kappa*vbar*tau/gamma/gamma *(kappa-gamma*rho*i*u-D1(u))\
        - 2*kappa*vbar/gamma/gamma*np.log((1.0-g(u)*np.exp(-D1(u)*tau))/(1.0-g(u)))
    # Characteristic function for the Heston's model    
    cf = lambda u: np.exp(A(u) + C(u)*v0)
    return cf 

def EUOptionPriceFromMCPathsGeneralized(CP,S,K,T,r):
    # S is a vector of Monte Carlo samples at T
    result = np.zeros([len(K),1])
    if CP == OptionType.CALL:
        for (idx,k) in enumerate(K):
            result[idx] = np.exp(-r*T)*np.mean(np.maximum(S-k,0.0))
    elif CP == OptionType.PUT:
        for (idx,k) in enumerate(K):
            result[idx] = np.exp(-r*T)*np.mean(np.maximum(k-S,0.0))
    return result

def GeneratePathsHestonEuler(NoOfPaths,NoOfSteps,T,r,S_0,kappa,gamma,rho,vbar,v0):    
    Z1 = np.random.normal(0.0,1.0,[NoOfPaths,NoOfSteps])
    Z2 = np.random.normal(0.0,1.0,[NoOfPaths,NoOfSteps])
    W1 = np.zeros([NoOfPaths, NoOfSteps+1])
    W2 = np.zeros([NoOfPaths, NoOfSteps+1])
    V = np.zeros([NoOfPaths, NoOfSteps+1])
    X = np.zeros([NoOfPaths, NoOfSteps+1])
    V[:,0]=v0
    X[:,0]=np.log(S_0)
    
    time = np.zeros([NoOfSteps+1])
        
    dt = T / float(NoOfSteps)
    for i in range(0,NoOfSteps):
        # making sure that samples from normal have mean 0 and variance 1
        if NoOfPaths > 1:
            Z1[:,i] = (Z1[:,i] - np.mean(Z1[:,i])) / np.std(Z1[:,i])
            Z2[:,i] = (Z2[:,i] - np.mean(Z2[:,i])) / np.std(Z2[:,i])
        Z2[:,i] = rho * Z1[:,i] + np.sqrt(1.0-rho**2)*Z2[:,i]
        
        W1[:,i+1] = W1[:,i] + np.power(dt, 0.5)*Z1[:,i]
        W2[:,i+1] = W2[:,i] + np.power(dt, 0.5)*Z2[:,i]
        
        # Truncated boundary condition
        V[:,i+1] = V[:,i] + kappa*(vbar - V[:,i]) * dt + gamma* np.sqrt(V[:,i]) * (W1[:,i+1]-W1[:,i])
        V[:,i+1] = np.maximum(V[:,i+1],0.0)
        
        X[:,i+1] = X[:,i] + (r - 0.5*V[:,i])*dt + np.sqrt(V[:,i])*(W2[:,i+1]-W2[:,i])
        time[i+1] = time[i] +dt
        
    #Compute exponent
    S = np.exp(X)
    paths = {"time":time,"S":S}
    return paths

def CIR_Sample(NoOfPaths,kappa,gamma,vbar,s,t,v_s):
    delta = 4.0 *kappa*vbar/gamma/gamma
    c= 1.0/(4.0*kappa)*gamma*gamma*(1.0-np.exp(-kappa*(t-s)))
    kappaBar = 4.0*kappa*v_s*np.exp(-kappa*(t-s))/(gamma*gamma*(1.0-np.exp(-kappa*(t-s))))
    sample = c* np.random.noncentral_chisquare(delta,kappaBar,NoOfPaths)
    return  sample

def GeneratePathsHestonAES(NoOfPaths,NoOfSteps,T,r,S_0,kappa,gamma,rho,vbar,v0):    
    Z1 = np.random.normal(0.0,1.0,[NoOfPaths,NoOfSteps])
    W1 = np.zeros([NoOfPaths, NoOfSteps+1])
    V = np.zeros([NoOfPaths, NoOfSteps+1])
    X = np.zeros([NoOfPaths, NoOfSteps+1])
    V[:,0]=v0
    X[:,0]=np.log(S_0)
    
    time = np.zeros([NoOfSteps+1])
        
    dt = T / float(NoOfSteps)
    for i in range(0,NoOfSteps):
        # making sure that samples from normal have mean 0 and variance 1
        if NoOfPaths > 1:
            Z1[:,i] = (Z1[:,i] - np.mean(Z1[:,i])) / np.std(Z1[:,i])
        W1[:,i+1] = W1[:,i] + np.power(dt, 0.5)*Z1[:,i]
        
        # Exact samles for the variance process
        V[:,i+1] = CIR_Sample(NoOfPaths,kappa,gamma,vbar,0,dt,V[:,i])
        k0 = (r -rho/gamma*kappa*vbar)*dt
        k1 = (rho*kappa/gamma -0.5)*dt - rho/gamma
        k2 = rho / gamma
        X[:,i+1] = X[:,i] + k0 + k1*V[:,i] + k2 *V[:,i+1] + np.sqrt((1.0-rho**2)*V[:,i])*(W1[:,i+1]-W1[:,i])
        time[i+1] = time[i] +dt
        
    #Compute exponent
    S = np.exp(X)
    paths = {"time":time,"S":S}
    return paths

# Black-Scholes Call option price
def BS_Call_Put_Option_Price(CP,S_0,K,sigma,t,T,r):
    #print('Maturity T={0} and t={1}'.format(T,t))
    #print(float(sigma * np.sqrt(T-t)))
    #print('strike K ={0}'.format(K))
    K = np.array(K).reshape([len(K),1])
    d1    = (np.log(S_0 / K) + (r + 0.5 * np.power(sigma,2.0)) 
    * (T-t)) / (sigma * np.sqrt(T-t))
    d2    = d1 - sigma * np.sqrt(T-t)
    if CP == OptionType.CALL:
        value = st.norm.cdf(d1) * S_0 - st.norm.cdf(d2) * K * np.exp(-r * (T-t))
     #   print(value)
    elif CP == OptionType.PUT:
        value = st.norm.cdf(-d2) * K * np.exp(-r * (T-t)) - st.norm.cdf(-d1)*S_0
      #  print(value)
    return value

def mainCalculation():
    NoOfPaths = 2500
    NoOfSteps = 500
    
    # Heston model parameters
    gamma = 1.0
    kappa = 0.5
    vbar  = 0.04
    rho   = -0.9
    v0    = 0.04
    T     = 1.0
    S_0   = 100.0
    r     = 0.1
    CP    = OptionType.CALL
    
    # First we define a range of strikes and check the convergence
    K = np.linspace(80,S_0*1.5,30)
    
    # Exact solution with the COS method
    cf = ChFHestonModel(r,T,kappa,gamma,vbar,v0,rho)
    
    # The COS method
    optValueExact = CallPutOptionPriceCOSMthd(cf, CP, S_0, r, T, K, 1000, 8)
    
    # Euler simulation
    pathsEULER = GeneratePathsHestonEuler(NoOfPaths,NoOfSteps,T,r,S_0,kappa,gamma,rho,vbar,v0)
    S_Euler = pathsEULER["S"]
    
    # Almost exact simulation
    pathsAES = GeneratePathsHestonAES(NoOfPaths,NoOfSteps,T,r,S_0,kappa,gamma,rho,vbar,v0)
    S_AES = pathsAES["S"]
    
    # S_Euler[:,-1] >> every path @ last time T     
    OptPrice_EULER = EUOptionPriceFromMCPathsGeneralized(CP,S_Euler[:,-1],K,T,r)
    OptPrice_AES   = EUOptionPriceFromMCPathsGeneralized(CP,S_AES[:,-1],K,T,r)
    
    plt.figure(1)
    plt.plot(K,optValueExact,'-r')
    plt.plot(K,OptPrice_EULER,'--k')
    plt.plot(K,OptPrice_AES,'.b')
    plt.legend(['Exact (COS)','Euler','AES'])
    plt.grid()
    plt.xlabel('strike, K')
    plt.ylabel('option price')
    
    # Here we will analyze the convergence for particular dt
    dtV = np.array([1.0, 1.0/4.0, 1.0/8.0,1.0/16.0,1.0/32.0,1.0/64.0])
    NoOfStepsV = [int(T/x) for x in dtV]
    
    # Specify strike for analysis
    K = np.array([100.0])
    
    # Exact
    optValueExact = CallPutOptionPriceCOSMthd(cf, CP, S_0, r, T, K, 1000, 8)
    errorEuler = np.zeros([len(dtV),1])
    errorAES = np.zeros([len(dtV),1])
    
    for (idx,NoOfSteps) in enumerate(NoOfStepsV):
        # Euler
        np.random.seed(3)
        pathsEULER = GeneratePathsHestonEuler(NoOfPaths,NoOfSteps,T,r,S_0,kappa,gamma,rho,vbar,v0)
        S_Euler = pathsEULER["S"]
        OptPriceEULER = EUOptionPriceFromMCPathsGeneralized(CP,S_Euler[:,-1],K,T,r)
        errorEuler[idx] = OptPriceEULER-optValueExact
        # AES
        np.random.seed(3)
        pathsAES = GeneratePathsHestonAES(NoOfPaths,NoOfSteps,T,r,S_0,kappa,gamma,rho,vbar,v0)
        S_AES = pathsAES["S"]
        OptPriceAES   = EUOptionPriceFromMCPathsGeneralized(CP,S_AES[:,-1],K,T,r)
        errorAES[idx] = OptPriceAES-optValueExact
    
    # Print the results
    for i in range(0,len(NoOfStepsV)):
        print("Euler Scheme, K ={0}, dt = {1} = {2}".format(K,dtV[i],errorEuler[i]))
        
    for i in range(0,len(NoOfStepsV)):
        print("AES Scheme, K ={0}, dt = {1} = {2}".format(K,dtV[i],errorAES[i]))


mainCalculation()
```
